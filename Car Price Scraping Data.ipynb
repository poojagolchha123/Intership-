```python
driver=webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
```

```python
driver.get('https://www.olx.in/')
```

```python
search=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div/div[1]/div[2]/div[1]/a/span')
search.click()
```

```python
name=[]

for i in driver.find_elements_by_xpath('//div[@class="_4aNdc"]'):
    name.append(i.text)
name    
```

```python
km=[]

for i in driver.find_elements_by_xpath('//div[@class="KFHpP"]'):
    km.append(i.text)
    
km
```

```python
price=[]

for i in driver.find_elements_by_xpath('//span[@class="_3GOwr"]'):
    price.append(i.text)
    
price
```

```python
df=pd.DataFrame({'Name':name,"Year-Km":km,"Price":price})
```
