# 1 Scrape data for 'Data Analyst'Job position in "Banglore"

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver=webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element_by_class_name("suggestor-input")
designation.send_keys("Data Analyst")

location=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
location.send_keys("Banglore")

search=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]")
search.click()
```

```python
job=[]
company=[]
location=[]
experience=[]
```

```python
for i in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    job.append(i.text)
    
job

for i in driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]'):
    company.append(i.text)
company

for i in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]/span[1]'):
    experience.append(i.text)
experience

for i in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]'):
    location.append(i.text)
    
location

print(len(job),len(company),len(experience),len(location))
df=pd.DataFrame({'Jobs':job,"Company":company,"Experience":experience,"location":location})
df
```

# 10)Write a python program to scrape data for Data Scientist designation

```python
driver=webdriver.Chrome(r'C:\Web Driver\chromedriver.exe')
driver.get('https://www.ambitionbox.com/')
```

```python
salary=driver.find_element_by_xpath("/html/body/div[1]/nav/nav/a[4]")
salary.click()
```

```python
designation=[]

for i in driver.find_elements_by_xpath('//p[@class="jp"]'):
    designation.append(i.text)
    
designation
```

```python
salary=[]

for i in driver.find_elements_by_xpath('//div[@class="result-col salary-range"]'):
    salary.append(i.text.split())
salary
```
